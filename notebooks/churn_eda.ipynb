```python
import pandas as pd
```

```python
test_data  = pd.read_csv('../data/customer_churn_dataset-testing-master.csv')
train_data = pd.read_csv('../data/customer_churn_dataset-training-master.csv')
```

```python
print(train_data.shape)
print(test_data.shape)
train_data.head()
train_data.info()
train_data['Churn'].value_counts()
```

```python
print(train_data['Gender'].unique())
print(train_data['Subscription Type'].unique())
print(train_data['Contract Length'].unique())
```

```python
train_data.describe()
```

```python
train_data = train_data.dropna(axis = 0)
train_data = train_data.drop(columns=['CustomerID'])
train_data.head()
```

```python
train_data['Gender'] = train_data['Gender'].map({'Male':1,'Female':0})
train_data = pd.get_dummies(train_data, columns=['Subscription Type','Contract Length'],drop_first= True) 
train_X = train_data.drop(columns=['Churn'])
train_y = train_data['Churn']
```

```python
train_data.head()
#train_data.columns.to_list()
```

```python
test_data.head()
```

```python
test_data = test_data.dropna(axis=0)
test_data = test_data.drop(columns= ['CustomerID'])
test_data['Gender'] = test_data['Gender'].map({'Male':1, 'Female':0})
test_data = pd.get_dummies(test_data, columns=['Subscription Type', 'Contract Length'], drop_first= True)
```

```python
test_X = test_data.drop(columns=['Churn'])
test_y = test_data['Churn']
```

```python
print(train_X.columns.to_list())
print(test_X.columns.to_list())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_X_scaled = scaler.fit_transform(train_X)
test_X_scaled  = scaler.transform(test_X)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=100)
log_model.fit(train_X_scaled,train_y)
log_preds = log_model.predict(test_X_scaled)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy Test: {}\n".format(accuracy_score(log_preds, test_y)))
print("Classification Report: \n{}\n".format(classification_report(test_y, log_preds)))
print("Confusion matrix: \n {}\n".format(confusion_matrix(test_y,log_preds)))
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier(random_state= 1, max_depth= 5, min_samples_leaf=50)
tree_model.fit(train_X,train_y)
tree_preds = tree_model.predict(test_X)
```

```python
print("Accuracy Score: {}".format(accuracy_score(test_y,tree_preds)))
print("Classification Report: \n {}".format(classification_report(test_y,tree_preds)))
print("Confusion Matrix: \n{}".format(confusion_matrix(test_y,tree_preds)))
```

```python
train_pred = tree_model.predict(train_X)
print("Training accuracy: ", accuracy_score(train_y, train_pred))
```

```python
print("Correlation with Churn (train):")
print(train_data.corr(numeric_only=True)['Churn'].sort_values(ascending=False))

print("\nCorrelation with Churn (test):")
print(test_data.corr(numeric_only=True)['Churn'].sort_values(ascending=False))
```

```python
combined_df = pd.concat([train_data, test_data], ignore_index=True)

from sklearn.model_selection import train_test_split
X = combined_df.drop(columns=['Churn'])
y = combined_df['Churn']

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest_model = RandomForestClassifier(n_estimators=100, random_state=1)
forest_model.fit(train_X,train_y)
forest_predict = forest_model.predict(test_X) 
```

```python
print("Accuracy: ", accuracy_score(test_y, forest_predict))
print("Classification Report: \n",classification_report(test_y,forest_predict))
print("Confusion Matrix: \n", confusion_matrix(test_y,forest_predict))
```

```python
importances = pd.Series(forest_model.feature_importances_, index=train_X.columns)
importances = importances.sort_values(ascending=False)
print(importances)
```
